(b) Check if a number is prime or not using function

```python
# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # If divisible, not a prime number
    return True  # If no divisors, it's prime

# Use the function to check numbers
num = 18
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
#Euler method:

import numpy as np
import matplotlib.pyplot as plt

# Function for f(x) = 4 - x^2
def f(x):
    return 4 - x**2

# Euler's Method
def euler_method(f, x0, I0, h, x_end):
    x_values = np.arange(x0, x_end+h, h)
    I_values = np.zeros(len(x_values))
    I_values[0] = I0

    for i in range(1, len(x_values)):
        I_values[i] = I_values[i-1] + h * f(x_values[i-1])

    return x_values, I_values

  # Parameters
x0 = 0      # Start point
I0 = 0      # Initial value
h = 0.2     # Step size
x_end = 2   # End point
exact_solution = (8 - 8/3)  # 16/3

# Compute numerical solutions
x_euler, I_euler = euler_method(f, x0, I0, h, x_end)

# Print final results
print(f"Euler Method: I({x_end}) â‰ˆ {I_euler[-1]:.6f}")
print(f"Exact Solution: I({x_end}) = {exact_solution:.6f}")

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(x_euler, I_euler, 'bo-', label="Euler Method")
plt.axhline(y=exact_solution, color='k', linestyle='--', label="Exact Solution")
plt.xlabel("x")
plt.ylabel("Integral Value I(x)")
plt.title("Numerical Integration using Euler, Trapezoidal, and RK4")
plt.legend()
plt.grid()
plt.show()
```

## **Use the similar approch using the methods Trapezoidal and Runge-Kutta**

```python
#function f(x)=2-x**2

import numpy as np
import matplotlib.pyplot as plt

# Function for f(x) = 4 - x^2
def f(x):
    return 4 - x**2
# Euler's Method
def euler_method(f, x0, I0, h, x_end):
    x_values = np.arange(x0, x_end+h, h)
    I_values = np.zeros(len(x_values))
    I_values[0] = I0

    for i in range(1, len(x_values)):
        I_values[i] = I_values[i-1] + h * f(x_values[i-1])

    return x_values, I_values
# Trapezoidal Method
def trapezoidal_method(f, x0, I0, h, x_end):
    x_values = np.arange(x0, x_end+h, h)
    I_values = np.zeros(len(x_values))
    I_values[0] = I0

    for i in range(1, len(x_values)):
        I_values[i] = I_values[i-1] + (h / 2) * (f(x_values[i-1]) + f(x_values[i]))

    return x_values, I_values

# Runge-Kutta 4th Order Method
def runge_kutta_4(f, x0, I0, h, x_end):
    x_values = np.arange(x0, x_end+h, h)
    I_values = np.zeros(len(x_values))
    I_values[0] = I0


    for i in range(1, len(x_values)):
        x = x_values[i-1]
        k1 = h * f(x)
        k2 = h * f(x + h/2)
        k3 = h * f(x + h/2)
        k4 = h * f(x + h)
        I_values[i] = I_values[i-1] + (1/6) * (k1 + 2*k2 + 2*k3 + k4)

    return x_values, I_values

# Parameters
x0 = 0      # Start point
I0 = 0      # Initial value
h = 0.2     # Step size
x_end = 2   # End point
exact_solution = (8 - 8/3)  # 16/3

# Compute numerical solutions
x_euler, I_euler = euler_method(f, x0, I0, h, x_end)
x_trap, I_trap = trapezoidal_method(f, x0, I0, h, x_end)
x_rk4, I_rk4 = runge_kutta_4(f, x0, I0, h, x_end)

# Print final results
print(f"Euler Method: I({x_end}) â‰ˆ {I_euler[-1]:.6f}")
print(f"Trapezoidal Method: I({x_end}) â‰ˆ {I_trap[-1]:.6f}")
print(f"Runge-Kutta Method: I({x_end}) â‰ˆ {I_rk4[-1]:.6f}")
print(f"Exact Solution: I({x_end}) = {exact_solution:.6f}")

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(x_euler, I_euler, 'bo-', label="Euler Method")
plt.plot(x_trap, I_trap, 'ro-', label="Trapezoidal Method")
plt.plot(x_rk4, I_rk4, 'go-', label="Runge-Kutta 4")
plt.axhline(y=exact_solution, color='k', linestyle='--', label="Exact Solution")
plt.xlabel("x")
plt.ylabel("Integral Value I(x)")
plt.title("Numerical Integration using Euler, Trapezoidal, and RK4")
plt.legend()
plt.grid()
plt.show()
```

## **Integration by Monte Carlo method**

Monte-Carlo-integration.jpg

```python
import numpy as np

# Define the function to integrate
def f(x):
    return np.exp(x)

# Monte Carlo Integration function
def monte_carlo_integration(f, a, b, num_samples):
    x_samples = np.random.uniform(a, b, num_samples)  # Generate random x values
    f_values = f(x_samples)  # Evaluate f(x) at random points
    integral = (b - a) * np.mean(f_values)  # Compute integral estimate
    return integral

# Parameters
a, b = 0, 1  # Integration limits
num_samples = 10000  # Number of random points

# Compute integral using Monte Carlo
estimated_integral = monte_carlo_integration(f, a, b, num_samples)

# Exact solution for comparison
exact_integral = np.exp(1) - 1

# Print results
print(f"Monte Carlo Estimated Integral: {estimated_integral:.6f}")
print(f"Exact Integral: {exact_integral:.6f}")
print(f"Error: {abs(estimated_integral - exact_integral):.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return np.exp(x)

# Monte Carlo Integration function
def monte_carlo_integration(f, a, b, num_samples):
    x_samples = np.random.uniform(a, b, num_samples)  # Generate random x values
    f_values = f(x_samples)  # Evaluate f(x) at random points
    integral = (b - a) * np.mean(f_values)  # Compute integral estimate
    return integral

# Parameters
a, b = 0, 1  # Integration limits
sample_sizes = [100, 1000, 10000, 100000, 1000000, 10000000, 100000000]  # Different sample sizes
exact_integral = np.exp(1) - 1  # Exact solution
estimated_integrals = []  # Store estimated values

# Compute integral for different sample sizes
for N in sample_sizes:
    estimate = monte_carlo_integration(f, a, b, N)
    estimated_integrals.append(estimate)
    print(f"Samples: {N}, Estimated Integral: {estimate:.6f}, Error: {abs(estimate - exact_integral):.6f}")

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, estimated_integrals, 'bo-', label="Monte Carlo Estimate")
plt.axhline(y=exact_integral, color='r', linestyle='--', label="Exact Integral")
plt.xscale("log")  # Log scale for better visualization
plt.xlabel("Number of Samples (log scale)")
plt.ylabel("Estimated Integral Value")
plt.title("Monte Carlo Integration vs. Exact Integral")
plt.legend()
plt.grid()
plt.show()

```

## *Estimation* of the value of pi

Monte-Carlo-pi.jpg

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to estimate Pi using Monte Carlo
def monte_carlo_pi(num_samples):
    x = np.random.uniform(0, 1, num_samples)  # Random x-coordinates
    y = np.random.uniform(0, 1, num_samples)  # Random y-coordinates

    # Check if points are inside the quarter-circle (x^2 + y^2 <= 1)
    inside_circle = x**2 + y**2 <= 1

    # Estimate Pi
    pi_estimate = 4 * np.sum(inside_circle) / num_samples

    return x, y, inside_circle, pi_estimate

# Sample sizes
num_samples_list = [100, 1000, 10000, 100000]  # Increasing number of samples

# Plot for different sample sizes
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for i, num_samples in enumerate(num_samples_list):
    x, y, inside_circle, pi_estimate = monte_carlo_pi(num_samples)

    ax = axes[i // 2, i % 2]  # Select subplot
    ax.scatter(x[inside_circle], y[inside_circle], color="blue", s=1, label="Inside Circle")
    ax.scatter(x[~inside_circle], y[~inside_circle], color="red", s=1, label="Outside Circle")

    ax.set_title(f"Samples: {num_samples}, Estimated Ï€: {pi_estimate:.6f}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.legend()

plt.suptitle("Monte Carlo Estimation of Ï€ with Increasing Samples")
plt.tight_layout()
plt.show()
```

## **Solving Equations**
