A better plot

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Data for the free energy profile
G = [G1, G2, G3, G4, G5]  # Assuming G1, G2, G3, G4, G5 are defined
E = [E1, E2, E3, E4, E5]
S = [S1, S2, S3, S4, S5]
steps = np.array([1, 2, 3, 4, 5])

# Define box properties
box_width = 0.4  # Wide rectangles
box_height = 0.1  # Low height
gap = 0.04  # Small gap to ensure lines don't touch boxes

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Function to draw rectangles and segmented lines
def plot_energy_profile(steps, values, color, label):
    for i in range(len(steps)):
        rect = Rectangle((steps[i] - box_width / 2, values[i] - box_height / 2),
                         box_width, box_height, edgecolor='black',
                         facecolor=color, linestyle='solid', joinstyle='round')
        ax.add_patch(rect)

    # Draw dotted lines between rectangles without touching them
    for i in range(len(steps) - 1):
        x_start = steps[i] + box_width / 2 + gap
        x_end = steps[i + 1] - box_width / 2 - gap
        ax.plot([x_start, x_end], [values[i], values[i + 1]], linestyle='dotted',
                color=color, linewidth=2)

# Plot data with modified markers
plot_energy_profile(steps, G, 'blue', 'G')
plot_energy_profile(steps, E, 'red', 'E')
plot_energy_profile(steps, S, 'green', 'S')

# Customize the plot
ax.set_xlabel('Reaction Steps', fontsize=14)
ax.set_ylabel('Free Energy (kcal/mol)', fontsize=14)
ax.set_title('Free Energy Profile', fontsize=16)

# Ensure all steps are shown on the x-axis
ax.set_xticks(steps)

# Add grid for better visualization
ax.grid(True)

# Show the plot
plt.show()
```

## Integrate 'p_z_total' cosidering it Y and 'energy' considering as X till X=-infinity to X=0

```python

import numpy as np

# Find the index where energy is closest to zero
energy_index = (np.abs(pdos['energy'] - 0)).argmin()

# Method 2: Using Trapezoidal rule
# Create a function to perform Trapezoidal rule integration
def integrate_trapz(energy, p_z_total):
    integral_value = np.trapz(p_z_total, energy)
    return integral_value

# Integrate p_z_total with respect to energy up to energy=0
integral_value_trapz = integrate_trapz(pdos['energy'][:energy_index], pdos['p_z_total'][:energy_index])

print(f"The integral of p_z_total (Trapezoidal Rule) is: {integral_value_trapz}")
```

## plot  GOH vs D(EF) and do the linear fitting

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Assuming 'des' DataFrame is already loaded and contains columns 'GOH' and 'D(EF)'
# Replace 'GOH' and 'D(EF)' with the actual column names if they are different

# Extract the data
x = des['D(EF)']
y = des['GOH']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Calculate R-squared
r_squared = r_value**2

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data Points')

# Plot the regression line
plt.plot(x, slope * x + intercept, color='red', label='Linear Fit')

# Add labels and title
plt.xlabel('D(EF)')
plt.ylabel('GOH')
plt.title('GOH vs D(EF) with Linear Fit')

# Add R-squared and equation of the line to the plot
equation = f'y = {slope:.2f}x + {intercept:.2f}'
plt.text(0.05, 0.95, f'$R^2$ = {r_squared:.3f}\n{equation}', transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top')


# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print additional details
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_squared:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Standard Error: {std_err:.3f}")
```
